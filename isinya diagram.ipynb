```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Menghitung jumlah setiap kategori dalam 'smoking_status'
smoking_counts = df['smoking_status'].value_counts()

# Menampilkan tabel distribusi
print(smoking_counts)
```

```python
# Membuat pie chart
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Menampilkan tabel
ax[0].axis('off')  # Menyembunyikan axis
ax[0].table(cellText=smoking_counts.reset_index().values, 
            colLabels=['Smoking Status', 'Count'], 
            loc='center')

# Membuat pie chart
ax[1].pie(smoking_counts, labels=smoking_counts.index, autopct='%1.2f%%', startangle=90)
ax[1].axis('equal')  # Membuat pie chart berbentuk lingkaran

# Menampilkan hasil
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset CSV
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Mengelompokkan data berdasarkan 'smoking_status' dan 'stroke', kemudian menghitung jumlahnya
grouped_data = df.groupby(['smoking_status', 'stroke']).size().unstack(fill_value=0)

# Menampilkan hasil pengelompokkan
print(grouped_data)

# Membuat diagram batang (bar plot)
grouped_data.plot(kind='bar', stacked=False, figsize=(10, 6), color=['skyblue', 'salmon'])

# Menambahkan label dan judul
plt.title('Distribusi Stroke berdasarkan Status Merokok')
plt.xlabel('Status Merokok')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=0)  # Memiringkan label sumbu X
plt.legend(title='Stroke', labels=['Tidak Stroke (0)', 'Stroke (1)'])

# Menampilkan diagram
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset CSV
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Mengelompokkan data berdasarkan 'smoking_status', 'gender', dan 'stroke' dan menghitung jumlahnya
grouped_data = df.groupby(['smoking_status', 'gender', 'stroke']).size().unstack(fill_value=0)

# Menampilkan hasil pengelompokkan
print(grouped_data)

# Membuat diagram batang bertumpuk
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 7), color=['skyblue', 'salmon'])

# Menambahkan label dan judul
plt.title('Distribusi Stroke Berdasarkan Status Merokok, Gender, dan Umur')
plt.xlabel('Status Merokok dan Gender')
plt.ylabel('Jumlah Kasus')
plt.xticks(rotation=45, ha='right')  # Memiringkan label sumbu X agar mudah dibaca
plt.legend(title='Stroke', labels=['Tidak Stroke (0)', 'Stroke (1)'])

# Menampilkan diagram
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset CSV
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Menghitung jumlah setiap kategori dalam kolom 'ever_married'
ever_married_counts = df['ever_married'].value_counts()

# Membuat pie chart dan tabel dalam satu plot
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Menampilkan tabel
ax[0].axis('off')  # Menyembunyikan axis
ax[0].table(cellText=ever_married_counts.reset_index().values, 
            colLabels=['Ever Married', 'Count'], 
            loc='center')

# Membuat pie chart
ax[1].pie(ever_married_counts, labels=ever_married_counts.index, autopct='%1.2f%%', startangle=90)
ax[1].axis('equal')  # Membuat pie chart berbentuk lingkaran

# Menampilkan hasil
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset CSV
df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Membuat histogram dua dimensi antara 'stroke' dan 'age'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='bmi', hue='stroke', multiple="stack", kde=False, bins=30, palette='coolwarm')

# Menambahkan label dan judul
plt.title('Histogram Dua Dimensi antara Stroke dan bmi')
plt.xlabel('Umur')
plt.ylabel('Jumlah Kasus')

# Menampilkan histogram
plt.show()
```
